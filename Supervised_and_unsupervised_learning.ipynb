#3. Data Preprocessing

At this stage, the Digits dataset is divided into two parts: training data and test data, with a ratio of 80:20. This division is carried out to ensure that the model learns from the training data and is evaluated using the test data to objectively measure its performance.

###3.1 Data Spliting

```python
from sklearn.model_selection import train_test_split  #For split the dataset into training and testing data

# Membagi data menjadi train dan test
x_train, x_test, y_train, y_test = train_test_split(df_x, df_y, test_size=0.2, random_state=42)
```

# 4. Machine Learning Models
Machine learning models are algorithms that learn patterns from data and make predictions. In this project, we use different classification models to recognize handwritten digits. These models include:

1. Logistic Regression: A statistical model used for binary and multi-class classification.

2. K-Nearest Neighbors (KNN): A model that classifies data based on the nearest neighbors.

3. Support Vector Machine (SVM): A model that separates data using an optimal margin.

4. Random Forest: A model that combines multiple decision trees to improve accuracy.

5. Neural Network (MLPClassifier): A model that mimics the human brain using multiple layers of neurons.

Each model is trained using the training data and evaluated using the test data. We use accuracy, a confusion matrix, and a classification report to measure performance.


###4.1 Define Models

```python
#Import library models
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix  # For model evaluation
from sklearn.linear_model import LogisticRegression  # For logistic regression model
from sklearn.neighbors import KNeighborsClassifier  # For K-Nearest Neighbors (KNN) model
from sklearn.svm import SVC  # For Support Vector Classifier (SVC) model
from sklearn.ensemble import RandomForestClassifier  # For Random Forest model
from sklearn.neural_network import MLPClassifier  # For Multi-layer Perceptron (MLP) model

# Define models
models = {
    "Logistic Regression": LogisticRegression(max_iter=5000),
    "KNN": KNeighborsClassifier(n_neighbors=3),
    "SVM": SVC(kernel='rbf', C=1, gamma=0.001),
    "Random Forest": RandomForestClassifier(n_estimators=100),
    "MLP (Neural Network)": MLPClassifier(hidden_layer_sizes=(100,), max_iter=500)
}
```

###4.2 Train and Validation Models

```python
results = {}
metrics = {}
for name, model in models.items():
    model.fit(x_train, y_train)  # Train the model
    y_pred = model.predict(x_test)  # Make predictions
    acc = accuracy_score(y_test, y_pred) # Calculate accuracy
    results[name] = acc  # Store accuracy
    print(f"{name} Accuracy: {acc:.4f}")

    report = classification_report(y_test, y_pred, output_dict=True) # Generate classification report
    metrics[name] = {
        "Accuracy": acc,
        "Precision": report["weighted avg"]["precision"],
        "Recall": report["weighted avg"]["recall"],
        "F1-Score": report["weighted avg"]["f1-score"]
    }

    # Classification Report
    print(classification_report(y_test, y_pred))

    # Confusion Matrix
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(5,5))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=digits.target_names, yticklabels=digits.target_names)
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.title(f"Confusion Matrix - {name}")
    plt.show()
```

# 6. Comparison Models


```python
# Convert metrics to DataFrame
metrics_df = pd.DataFrame(metrics).T  # Transpose
print(metrics_df)

# Plot  comparison Accuracy, Precision, Recall, F1-Score
plt.figure(figsize=(12, 6))
metrics_df.plot(kind="bar", figsize=(12, 6), colormap="coolwarm", edgecolor="black")
plt.title("Comparison of Model Performance", fontsize=14)
plt.ylabel("Score", fontsize=12)
plt.xticks(rotation=45, ha="right", fontsize=10)
plt.ylim(0.95, 1)  # Skala 0 - 1
plt.legend(title="Metrics", fontsize=10)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

#1. Import Library

```python
import pandas as pd  # For data manipulation and handling data structures
import numpy as np  # For numerical operations
import matplotlib.pyplot as plt  # For plotting data visualizations
import seaborn as sns  # For statistical data visualization
from sklearn.cluster import KMeans  # For the KMeans clustering algorithm
from sklearn.preprocessing import StandardScaler  # For data scaling (normalization)
from sklearn.metrics import silhouette_score  # To evaluate clustering performance
```

# 2. Load Dataset

```python
df = pd.read_csv('/content/Mall_Customers.csv')  # Read data from CSV into pandas DataFrame

# Display the first few rows of the dataset to understand its structure
print(df.head())  # Shows the first 5 rows of the dataset
```

# 4. Data Preprocessing

```python
# Select the features to be used for clustering (in this case: Age, Annual Income, and Spending Score)
features = df[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']]  # Choose relevant columns for clustering

# Scale the features using StandardScaler (important for KMeans clustering)
scaler = StandardScaler()  # Initialize the scaler
features_scaled = scaler.fit_transform(features)  # Scale the features and fit the scaler

# Display the first 5 rows of the scaled features to check the results of scaling
print(features_scaled[:5])  # Show first 5 rows of the scaled features
```

# 5. K-Means Clustering Model

```python
# Define the number of clusters (e.g., 5 clusters) for KMeans
kmeans = KMeans(n_clusters=5, random_state=42)  # Initialize the KMeans model with 5 clusters

# Fit the KMeans model to the scaled data
kmeans.fit(features_scaled)  # Apply the KMeans algorithm to the scaled data

# Add the cluster labels (result of clustering) to the original DataFrame
df['Cluster'] = kmeans.labels_  # Add the cluster labels to the dataframe

# Visualize the clustering result, with 'Age' on the x-axis and 'Annual Income' on the y-axis
plt.figure(figsize=(8, 6))  # Set the figure size for the plot
sns.scatterplot(x='Age', y='Annual Income (k$)', hue='Cluster', data=df, palette='Set1', s=100)  # Scatter plot with clusters color-coded
plt.title("Customer Segmentation Based on Age and Annual Income")  # Title of the plot
plt.show()  # Display the plot
```

# 7. Test the Model with New Data

```python
# Example: Predict the cluster for new customer data (Age=30, Income=40, Spending Score=60)
new_data = np.array([[30, 40, 60]])  # New customer data to be predicted

# Scale the new data using the same scaler that was used to scale the original features
new_data_scaled = scaler.transform(new_data)  # Apply the same scaling transformation to the new data

# Predict the cluster for the new data using the fitted KMeans model
new_cluster = kmeans.predict(new_data_scaled)  # Predict the cluster for the new data point

# Display the predicted cluster for the new customer
print(f"The new data belongs to cluster: {new_cluster[0]}")  # Print the predicted cluster for the new customer
```
